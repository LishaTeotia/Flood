```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
file_path = '/content/drive/My Drive/THESIS/flood_dataset.csv'
df = pd.read_csv(file_path)
a = np.array(df)
```

```python
df = df.dropna() # Drop no values
```

```python
X = df.drop(columns='Flood_Occurred')
Y = df['Flood_Occurred']
```

```python
# OneHotEncoding LULC
categorical_col = ['LULC_2020_Type', 'LULC_2021_Type', 'LULC_2022_Type', 'LULC_2023_Type']
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded = encoder.fit_transform(X[categorical_col])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_col))
X = X.drop(columns=categorical_col)
X = pd.concat([X.reset_index(drop=True), encoded_df], axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42) # 100 trees with max depth 5, 42 random no. generator seed for reproducibility
rf.fit(X_train, Y_train)
```

```python
Y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print (f"Accuracy: {accuracy:.2f}")
```

```python
c_report = classification_report(Y_test, Y_pred)
print("Classification report of RF model")
print(c_report)
```

```python
c_matrix = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:")
print(c_matrix)
```

```python
joblib.dump(rf, 'trained model path')
```
